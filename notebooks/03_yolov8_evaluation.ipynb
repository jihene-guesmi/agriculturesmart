```python
from ultralytics import YOLO

model = YOLO("../notebooks/runs/detect/pomegranate_detector-4/weights/best.pt")
```

```python
import os

test_dir = "../data/detection/test/images"

img_name = os.listdir(test_dir)[0]
img_path = os.path.join(test_dir, img_name)

results = model(img_path)
```

```python
#show result 
results[0].show()
```

```python
# test multiple images 
for i in range(5):
    img_name = os.listdir(test_dir)[i]
    img_path = os.path.join(test_dir, img_name)

    results = model(img_path)
```

```python
# show images  
import matplotlib.pyplot as plt

img_result = results[0].plot()

plt.imshow(img_result)
plt.axis("off")
plt.show()
```
